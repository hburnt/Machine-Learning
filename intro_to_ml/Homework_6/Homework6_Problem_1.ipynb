```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import torch.optim as optim
import pandas as pd
import numpy as np
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split

# Problem 1.a
url = 'https://raw.githubusercontent.com/hburnt/Intro-to-ML-datasets/main/Housing.csv'
df = pd.DataFrame(pd.read_csv(url))

var_list = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

def binary_map(input):
  return input.map({'no' : 0, 'yes' : 1})

df[var_list] = df[var_list].apply(binary_map)
df = df.drop('furnishingstatus', axis=1)

outputs = df.pop('price')
output = outputs.values

features = df.columns.tolist()
inputs = df[features].values

x_train, x_test, y_train, y_test = train_test_split (inputs, output, test_size = 0.2, random_state = 24);

# Standardizing Inputs
x_scaler = StandardScaler()
x_train_stand = x_scaler.fit_transform(x_train)
x_test_stand = x_scaler.fit_transform(x_test)

#Standardizing Outputs
y_scaler = StandardScaler()
y_train_stand = y_scaler.fit_transform(y_train.reshape(-1,1)).flatten()
y_test_stand = y_scaler.fit_transform(y_test.reshape(-1,1)).flatten()

# Convcerting to PyTorch Tensors
x_train_tensor = torch.tensor(x_train_stand, dtype=torch.float32)
x_test_tensor = torch.tensor(x_test_stand, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train_stand, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test_stand, dtype=torch.float32)

trainTensor = TensorDataset(x_train_tensor, y_train_tensor)
trainLoader = DataLoader(trainTensor, batch_size=32)

testTensor = TensorDataset(x_test_tensor, y_test_tensor)
testLoader = DataLoader(testTensor, batch_size =32)

model = nn.Sequential(
    nn.Linear(x_train.shape[1], 32),
    nn.Tanh(),
    nn.Linear(32, 1)
)

# Define loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.001)
epochs = 5000

for i in range(epochs +1):
    model.train()
    for inp, targ in trainLoader:
        optimizer.zero_grad()
        op = model(inp)
        loss = criterion(op.squeeze(), targ)
        loss.backward()
        optimizer.step()

    model.eval()
    with torch.no_grad():
        validation_op = model(x_test_tensor)
        validation_loss = criterion(validation_op.squeeze(), y_test_tensor)
        if i % 500 == 0:
            print(f'Epoch {i}, Training Loss: {loss.item()}, Validation Loss: {validation_loss.item()}')
```

```python
#Problem 1.b
df = pd.DataFrame(pd.read_csv(url))

var_list = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

def binary_map(input):
  return input.map({'no' : 0, 'yes' : 1})

df[var_list] = df[var_list].apply(binary_map)
df = df.drop('furnishingstatus', axis=1)

outputs = df.pop('price')
output = outputs.values

features = df.columns.tolist()
inputs = df[features].values

x_train, x_test, y_train, y_test = train_test_split (inputs, output, test_size = 0.2, random_state = 24);

# Standardizing Inputs
x_scaler = StandardScaler()
x_train_stand = x_scaler.fit_transform(x_train)
x_test_stand = x_scaler.fit_transform(x_test)

#Standardizing Outputs
y_scaler = StandardScaler()
y_train_stand = y_scaler.fit_transform(y_train.reshape(-1,1)).flatten()
y_test_stand = y_scaler.fit_transform(y_test.reshape(-1,1)).flatten()

# Convcerting to PyTorch Tensors
x_train_tensor = torch.tensor(x_train_stand, dtype=torch.float32)
x_test_tensor = torch.tensor(x_test_stand, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train_stand, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test_stand, dtype=torch.float32)

trainTensor = TensorDataset(x_train_tensor, y_train_tensor)
trainLoader = DataLoader(trainTensor, batch_size=32)

testTensor = TensorDataset(x_test_tensor, y_test_tensor)
testLoader = DataLoader(testTensor, batch_size =32)

model = nn.Sequential(
    nn.Linear(x_train.shape[1], 32),
    nn.Tanh(),
    nn.Linear(32, 64),
    nn.Tanh(),
    nn.Linear(64,16),
    nn.Tanh(),
    nn.Linear(16, 1)
)

# Define loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.001)
epochs = 5000

for i in range(epochs +1):
    model.train()
    for inp, targ in trainLoader:
        optimizer.zero_grad()
        op = model(inp)
        loss = criterion(op.squeeze(), targ)
        loss.backward()
        optimizer.step()

    model.eval()
    with torch.no_grad():
        validation_op = model(x_test_tensor)
        validation_loss = criterion(validation_op.squeeze(), y_test_tensor)
        if i % 500 == 0:
            print(f'Epoch {i}, Training Loss: {loss.item()}, Validation Loss: {validation_loss.item()}')
```
